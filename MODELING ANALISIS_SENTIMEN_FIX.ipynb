## Preprocessing

### Case Folding

```python
def casefoldingText(text):
    if isinstance(text, str):  # Memeriksa apakah text adalah string
        return text.lower()   # Mengonversi ke huruf kecil
    else:
        return str(text)       # Mengembalikan nilai non-string seperti apa adanya
```

```python
data['Case Folding'] = data['full_text'].apply(casefoldingText)
data[['full_text', 'Case Folding']]
```

### Cleansing

CLEANSING
Cleansing adalah proses membersihkan dokumen dari kata yang tidak diperlukan. Kata-kata dibersihkan adalah kata-kata yang mengandung tweet entity seperti mention, retweet, hashtag, dan link url serta simbol atau kode karakter numeric (noisy text)

```python
def cleaningText(text):
    #menghapus tanda emoticon, chinese word, dll
    #text = text.encode('ascii', 'replace').decode('ascii')
    #menghapus tab, newline
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    #menghapus username
    text = re.sub('@[^\s]+', ' ', text)
    #menghapus link
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text)
    #menghapus tanda baca
    text = text.translate(str.maketrans(' ',' ',string.punctuation))
    #menghapus tanda baca javascript
    text = re.sub('<.*?>', ' ', text)
    #menghapus selain huruf alfabet (angka)
    text = re.sub('[^a-zA-Z]',' ', text)
    #menghapus single character
    text = re.sub(r"\b[a-zA-Z]\b"," ", text)
    text = ' '.join(text.split())
    return text
```

```python
data['Cleaning'] = data['Case Folding'].apply(cleaningText)
data[['Case Folding', 'Cleaning']]
```

Tokenizing

Hasil proses case folding, kalimat akan diproses dengan menguraikannya menjadi token-token atau kata-kata. Pada tahap ini akan dibantu dengan library NLTK.

```python
def tokenizingText(text):
    text = word_tokenize(text)
    return text
```

```python
data['Tokenizing'] = data['Cleaning'].apply(tokenizingText)
#data[['Cleaning', 'Tokenizing']]
```

### Normalisasi
Normalisasi data dalam analisis sentimen adalah proses untuk mengubah teks yang tidak terstruktur menjadi bentuk yang lebih terstruktur atau seragam

```python
slangWord = pd.read_excel('/content/drive/MyDrive/datadata/kamus_kata_alay.xlsx')

slangWord_dict = {}

for index, row in slangWord.iterrows():
    if row[0] not in slangWord_dict:
        slangWord_dict[row[0]] = row[1]

def slangWordText(text):
    text = [slangWord_dict[term] if term in slangWord_dict else term for term in text]
    return text
```

```python
data['Normalisasi'] = data['Tokenizing'].apply(slangWordText)
data[['Tokenizing', 'Normalisasi']]
```

### Stopword Removal

Hasil proses formalisasi kemudian akan dilakukan seleksi kata yang tidak penting dan menghapus kata tersebut. Tahap ini dibantu dengan library NLTK.

```python
stopword = stopwords.words('indonesian')
txt_stopword = pd.read_excel("/content/drive/MyDrive/datadata/stopwords_indo.xlsx", names=["stopwords"], header = None)

stopword.extend(["gacoan", "wah", "weh", "ih", "lol", "ya", "sok", "fess", "nya", "deh", "wkwk", "wkwkwkwkk", "wkwkwkw", "wuakakakak", "hehehe", "hmmmmm", "huftt", "hufty", "cmiiw", "huhu", "hmm", "lho", "dkk", "rl", "tuh", "sih", "nih", "teh", "kan", "atuh", "gas", "guys", "amp", "dll", "jg", "yg", "yaha", "the", "ia", "ada", "wes", "kok", "so", "okay", "dlu", "bu", "kak", "sih"])

stopword.extend(txt_stopword["stopwords"][0].split(' '))
stopword = set(stopword)

def stopwordRemoval(text):
    text = [word for word in text if word not in stopword]
    return text
```

```python
data['Stopword Removal'] = data['Normalisasi'].apply(stopwordRemoval)
data[['Normalisasi', 'Stopword Removal']]
```

### Stemming

Proses stemming dilakukan perubahan kata yang berimbuhan menjadi kata dasar. Tahap ini dibantu dengan library Sastrawi dan Swifter.

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in data['Stopword Removal']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])

print(term_dict)
print("------------------------")
```

```python
def get_stemmed_term(doc):
    return [term_dict[term] for term in doc]

data['Stemming'] = data['Stopword Removal'].apply(get_stemmed_term)
data[['Stopword Removal', 'Stemming']]
```

```python
# pelabelan sentimen menggunakan kamus inset (sumber: https://github.com/fajri91/InSet )
lexicon_positive = dict()
import csv
with open('/content/drive/MyDrive/datadata/lexicon_positive.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        lexicon_positive[row[0]] = int(row[1])

lexicon_negative = dict()
import csv
with open('/content/drive/MyDrive/datadata/lexicon_negative.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        lexicon_negative[row[0]] = int(row[1])

# Function to determine sentiment polarity of tweets
def sentiment_analysis_lexicon_indonesia(text):
    #for word in text:
    score = 0
    for word in text:
        if (word in lexicon_positive):
            score = score + lexicon_positive[word]
    for word in text:
        if (word in lexicon_negative):
            score = score + lexicon_negative[word]
    polarity=''
    if (score > 0):
        polarity = 'positif'
    elif (score < 0):
        polarity = 'negatif'
    else:
        polarity = 'netral'
    return score, polarity

results = data['Stemming'].apply(sentiment_analysis_lexicon_indonesia)
results = list(zip(*results))
data['jumlah_polarity'] = results[0]
data['sentimen'] = results[1]
print(data['sentimen'].value_counts())
```

## TF-IDF

```python
import difflib

review_tweet_list = data['Stemming'].tolist()

# Initialize the ratio variable with a low value
s = 0.0

# Iterate over the list and compare each pair of strings
for i in range(len(review_tweet_list)):
    for j in range(i + 1, len(review_tweet_list)):
        seq_matcher = difflib.SequenceMatcher(None, review_tweet_list[i], review_tweet_list[j])
        ratio = seq_matcher.ratio()
        if ratio > s:
            s = ratio

print("Jumlah RRatio:", s)
```

```python
def freq_words(x, terms = 30):
    all_words = ' '.join([text for text in x ])
    all_words = all_words.split()

    fdist = FreqDist(all_words)
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())})

    # Memlihi 10 Data Kata Tweet yang sering muncul
    d = words_df.nlargest(columns="count", n=terms)
    plt.figure(figsize=(25,7))
    ax = seaborn.barplot (data=d, x="word", y="count")
    ax.set(ylabel = 'Count')
    plt.show()
```

Naive Bayes Using K-Fold

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import classification_report

X = data['Stemming']  # Sesuaikan nama kolom sesuai dengan kolom fitur
y = data['sentimenlabel']  # Sesuaikan nama kolom sesuai dengan kolom target

X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert X_train and X_test to string format
X_train2 = [' '.join(map(str, document)) for document in X_train2]
X_test2 = [' '.join(map(str, document)) for document in X_test2]

# Apply TfidfVectorizer to all training and test data
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train2)
X_test_tfidf = vectorizer.transform(X_test2)

# Apply SMOTE for oversampling (Balancing Data SentimenLabel)
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_tfidf, y_train2)

# Train the classifier model
nb = MultinomialNB()
nb.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test data
y_pred2 = nb.predict(X_test_tfidf)

# Print the classification report
cm2= print(classification_report(y_test2, y_pred2))

# Define the scoring metrics for cross-validation
scoring = ['accuracy', 'recall_macro', 'precision_macro', 'f1_macro']

# Perform K-Fold Validation with multiple scoring metrics
cv_results = cross_validate(nb, X_train_resampled, y_train_resampled, cv=5, scoring=scoring)

# Print the results for each metric and each fold
for fold_idx, metrics in enumerate(zip(cv_results['test_accuracy'], cv_results['test_recall_macro'],
                                       cv_results['test_precision_macro'], cv_results['test_f1_macro'])):
    fold_accuracy, fold_recall, fold_precision, fold_f1 = metrics
    print(f"Fold {fold_idx + 1} - Accuracy: {fold_accuracy:.3f}, Recall: {fold_recall:.3f}, "
          f"Precision: {fold_precision:.3f}, F1-Score: {fold_f1:.3f}")

# Calculate and print the average scores across all folds
average_accuracy = cv_results['test_accuracy'].mean()
average_recall = cv_results['test_recall_macro'].mean()
average_precision = cv_results['test_precision_macro'].mean()
average_f1 = cv_results['test_f1_macro'].mean()

print(" ")
print("Average Accuracy:", average_accuracy)
print("Average Recall:", average_recall)
print("Average Precision:", average_precision)
print("Average F1-Score:", average_f1)
```

SVM Menggunakan K-FOLD VALIDATION

```python
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import KFold, train_test_split, cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_recall_fscore_support
from imblearn.over_sampling import SMOTE
import numpy as np

X = data['Stemming']  # Sesuaikan nama kolom sesuai dengan kolom fitur
y = data['sentimenlabel']  # Sesuaikan nama kolom sesuai dengan kolom target

# Split the data into training and testing sets (Training 70% Testing 30%)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.3, random_state=42)

# Ubah X_train dan X_test ke format string
X2 = [' '.join(map(str, document)) for document in X]

# Terapkan TfidfVectorizer ke semua data
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X2)

# Terapkan SMOTE untuk oversampling (Menyeimbangkan Data SentimenLabel)
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_tfidf, y)

# Tentukan k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Inisialisasi model SVM
svm = SVC(kernel='linear', random_state=42)

# Simpan nilai-nilai metrik untuk setiap fold
accuracies = []
recalls = []
precisions = []
f1_scores = []

# Lakukan cross-validation
for fold, (train_index, test_index) in enumerate(kfold.split(X_resampled), 1):
    X_train, X_test = X_resampled[train_index], X_resampled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    # Latih model SVM
    svm.fit(X_train, y_train)

    # Prediksi pada data uji
    y_pred_svm = svm.predict(X_test)

    # Hitung metrik
    accuracy = accuracy_score(y_test, y_pred_svm)
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred_svm, average='weighted')

    # Simpan nilai metrik untuk fold saat ini
    accuracies.append(accuracy)
    recalls.append(recall)
    precisions.append(precision)
    f1_scores.append(f1_score)

    # Cetak metrik untuk fold saat ini
    print(f"Fold {fold} - Accuracy: {accuracy:.3f}, Recall: {recall:.3f}, Precision: {precision:.3f}, F1-Score: {f1_score:.3f}")

# Hitung rata-rata metrik untuk semua fold
avg_accuracy = np.mean(accuracies)
avg_recall = np.mean(recalls)
avg_precision = np.mean(precisions)
avg_f1_score = np.mean(f1_scores)

# Cetak rata-rata metrik
print("\nAverage Metrics Across Folds:")
print(f"Average Accuracy: {avg_accuracy:.3f}")
print(f"Average Recall: {avg_recall:.3f}")
print(f"Average Precision: {avg_precision:.3f}")
print(f"Average F1-Score: {avg_f1_score:.3f}")

# Menghitung dan mencetak laporan klasifikasi untuk seluruh data menggunakan SVM
print("\nClassification Report for SVM:")
y_pred_svm_total = cross_val_predict(svm, X_resampled, y_resampled, cv=kfold)
print(classification_report(y_resampled, y_pred_svm_total))
```

### Random Forest Menggunakan K-FOLD VALIDATION

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report
import numpy as np

X = data['Stemming']  # Sesuaikan nama kolom sesuai dengan kolom fitur
y = data['sentimenlabel']  # Sesuaikan nama kolom sesuai dengan kolom target

# Split the data into training and testing sets (Training 70% Testing 30%)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.3, random_state=42)

# Konversi X_train dan X_test ke format string
X = [' '.join(map(str, document)) for document in X]

# Penerapan TfidfVectorizer untuk seluruh data
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)

# Terapkan SMOTE untuk oversampling (Balancing Data SentimenLabel)
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_tfidf, y)

# Inisialisasi model Random Forest
rf = RandomForestClassifier(random_state=42)

# Inisialisasi KFold dengan 5 lipatan
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Inisialisasi list untuk menyimpan metrik evaluasi
accuracy_scores = []
recall_scores = []
precision_scores = []
f1_scores = []

# Hitung dan simpan skor untuk setiap lipatan
for fold, (train_index, test_index) in enumerate(kfold.split(X_resampled), 1):
    # Tentukan data untuk lipatan saat ini
    X_train, X_test = X_resampled[train_index], X_resampled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    # Latih model pada data pelatihan
    rf.fit(X_train, y_train)

    # Lakukan prediksi pada data pengujian
    y_pred = rf.predict(X_test)

    # Hitung metrik evaluasi
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    # Tambahkan skor ke dalam list
    accuracy_scores.append(accuracy)
    recall_scores.append(recall)
    precision_scores.append(precision)
    f1_scores.append(f1)

    # Cetak hasil untuk setiap lipatan
    print(f"Fold {fold} - Accuracy: {accuracy:.3f}, Recall: {recall:.3f}, Precision: {precision:.3f}, F1-Score: {f1:.3f}")

# Hitung rata-rata skor dari semua lipatan
average_accuracy = np.mean(accuracy_scores)
average_recall = np.mean(recall_scores)
average_precision = np.mean(precision_scores)
average_f1 = np.mean(f1_scores)

# Cetak rata-rata skor
print("\nAverage Metrics Across Folds:")
print(f"Average Accuracy: {average_accuracy:.3f}")
print(f"Average Recall: {average_recall:.3f}")
print(f"Average Precision: {average_precision:.3f}")
print(f"Average F1-Score: {average_f1:.3f}")

# Hitung dan cetak laporan klasifikasi untuk seluruh data
print("\nClassification Report for Random Forest:")
print(classification_report(y_resampled, cross_val_predict(rf, X_resampled, y_resampled, cv=kfold)))
```
